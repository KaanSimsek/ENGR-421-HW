```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('hw01_data_points.csv', header=None)
labels = pd.read_csv('hw01_class_labels.csv', header=None)
```

```python
def train_test_split(data, labels):
    return data[:50000].values, labels[:50000].values.ravel(), data[50000:].values, labels[50000:].values.ravel()
```

```python
x_train, y_train, x_test, y_test = train_test_split(data, labels)
```

```python
print("x_train:",x_train.shape)
print("x_test:",x_test.shape)
print("y_train:",y_train.shape)
print("y_test:",y_test.shape)
```

```python
def estimate_prior_probabilities(y_train):
    unique_classes = np.unique(y_train)
    PR = [np.sum(y_train == c)/(y_train.size) for c in unique_classes]
    return PR
class_priors=estimate_prior_probabilities(y_train)
print(class_priors)
```

```python
def estimate_nucleotide_probabilities(x_train, y_train):
    unique_classes = np.unique(y_train)
    unique_x = np.unique(x_train)
    destinations = x_train.shape[1]

    indexs = x_train.shape[0]
    #res = [[[[(x_train[i][d] == x and y_train[i]==c) for d in range(0, destinations-1)] for i in range(0,indexs-1)] for c in unique_classes] for x in unique_x]
    #res = [[[np.sum(x_train[:][d] == x) / np.sum(y_train == c) for d in range(0,destinations)] for c in unique_classes] for x in unique_x]
    P=[]
     
    P =  [
            [
                [
                    sum(
                        x_train[i, d] == x and y_train[i] == c
                        for i in range(indexs)
                    ) / np.sum(y_train == c)
                    for d in range(destinations)
                ]
                for c in unique_classes
            ]
            for x in unique_x
        ]
    return P
        
P = estimate_nucleotide_probabilities(x_train, y_train)
print(P)
```

```python
def findIndex(arr, element):
    index=0
    for a in arr:
        if a == element:
            return index
        index += 1
    return -1
```

```python
def calculate_score_values(x, P, class_priors, unique_classes):
    unique_x = np.unique(x)
    destinations = x.shape[1]
    indexs = x.shape[0]
    
    score_list=[]
    for i in range(indexs):
        arr=[]
        for c in range(0,2):
            score=1
            for d in range(destinations):
                score = score * P[findIndex(unique_x, x[i][d])][c][d]
            score = np.log(score) + np.log(class_priors[c])
            arr.append(score)
        score_list.append(arr)
    return score_list
```

```python
unique_classes_train = np.unique(y_train)
score_list_train= calculate_score_values(x_train, P, class_priors, unique_classes_train)
print(score_list_train)
unique_classes_test = np.unique(y_test)
score_list_test= calculate_score_values(x_test, P, class_priors, unique_classes_test)
print(score_list_test)
```

```python
def calculate_confusion_matrix(y, scores):
    predicted_classes = np.argmax(scores, axis=1)
    predicted_classes= [p + 1 for p in predicted_classes]
    
    num_classes = len(np.unique(y))
    print(num_classes)
    confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)
    for true_class, predicted_class in zip(y, predicted_classes):
        confusion_matrix[predicted_class - 1, true_class - 1] += 1
    return confusion_matrix
```

```python
confusion_matrix_train = calculate_confusion_matrix(y_train, score_list_train)
confusion_matrix_test = calculate_confusion_matrix(y_test, score_list_test)
```
